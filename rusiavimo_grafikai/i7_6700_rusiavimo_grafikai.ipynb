```python
import pandas as pd
import os
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from IPython.display import display, Math, Latex
import math
%matplotlib inline
```

```python
failai={}
failai["Merge"]=pd.read_csv("6700_rezultatai/6700_mergesort.txt", sep=" ")
failai["Radix"]=pd.read_csv("6700_rezultatai/6700_radixsort.txt", sep=" ")
failai["Stable"]=pd.read_csv("6700_rezultatai/6700_stableSort.txt",sep=" ")
failai["SortSort"]=pd.read_csv("6700_rezultatai/6700_sort_elementai.txt",sep=" ")
failai["RadixByKey"]=pd.read_csv("6700_rezultatai/6700_radixSort_by_key_element.txt",sep=" ")
failai["SortSortByKey"]=pd.read_csv("6700_rezultatai/6700_sortbyelement.txt",sep=" ")
failai["StableByKey"]=pd.read_csv("6700_rezultatai/6700_stablesort_by_element.txt" ,sep=" ")

```

```python
plt.figure(figsize=(10,6))
for var in ["Merge","Radix","Stable","SortSort"]:
    plt.plot(failai[var]["N"],failai[var]["TIME"],label=var,linestyle="-",linewidth=3)
for var in ["SortSortByKey","StableByKey", "RadixByKey"]:
    plt.plot(failai[var]["N"],failai[var]["TIME"],label=var,linestyle="--",linewidth=3)
plt.legend()
plt.xlim(0,0.1e8)
plt.ylim(0,2.5)
plt.grid(color='grey', linestyle='-', linewidth=0.5)
plt.xlabel("Number of elements, N")
plt.ylabel("Time, s")
plt.show()


plt.figure(figsize=(10,6))
for var in ["Merge","Radix","Stable","SortSort"]:
    plt.plot(np.log2(failai[var]["N"]),(failai[var]["N"]/failai[var]["TIME"])/1000000.0,label=var,linestyle="-",linewidth=3)
for var in ["SortSortByKey","StableByKey", "RadixByKey"]:
    plt.plot(np.log2(failai[var]["N"]),(failai[var]["N"]/failai[var]["TIME"])/1000000.0,label=var,linestyle="--",linewidth=3)
plt.legend()

#plt.ylim(0,0.125)
plt.grid(color='grey', linestyle='-', linewidth=0.5)
plt.xlabel("Number of elements, lg(N)")
plt.ylabel("Millions per second, M/s")
plt.show()
```
